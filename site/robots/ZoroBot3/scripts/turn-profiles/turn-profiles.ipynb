# Turn Profiles for a Sinusoidal Angular Acceleration

## Setup

```python
%matplotlib inline
from matplotlib import pyplot
from micromouse_sinusoidal_turn_profiles_aux import Maze
from micromouse_sinusoidal_turn_profiles_aux import RobotPhysics
from micromouse_sinusoidal_turn_profiles_aux import Simulator
from micromouse_sinusoidal_turn_profiles_aux import pi


classic = Maze(cell=0.18, post=0.012)
half = Maze(cell=0.09, post=0.006)

robot = RobotPhysics(
    mass=0.070,
    moment_of_inertia=(0.070 * 0.05 ** 2) / 2,
    width=0.0702,
    wheels_separation=0.062,
    max_angular_velocity=40,
)

simulate = Simulator(robot=robot, maze=classic, time_period=0.00001)
```
